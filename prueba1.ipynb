```python
# Cargar datos desde archivo CSV local
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
# Mostrar primeras filas
df.sample(5)
```

```python
# se realiza la separacion de variables numericas y categoricas, para analisis
variables_numericas = ['Age', 'MonthlyIncome', 'DistanceFromHome', 'YearsAtCompany', 'JobSatisfaction']
variables_categoricas = ['Attrition', 'Department', 'Gender', 'MaritalStatus', 'OverTime']

print("\n--- TIPOS DE VARIABLES ---")
print("\nVariables numéricas seleccionadas:")
for col in variables_numericas:
    print(f"- {col}")

print("\nVariables categóricas seleccionadas:")
for col in variables_categoricas:
    print(f"- {col}")
```

```python
# analisis de variables categoricas de nuestra base
print("\n--- ANÁLISIS DE VARIABLES CATEGÓRICAS ---")

# Variable objetivo: Attrition
print("\nDistribución de Attrition (Variable objetivo):")
attrition_counts = df['Attrition'].value_counts()
attrition_percentage = df['Attrition'].value_counts(normalize=True) * 100
for valor, count in attrition_counts.items():
    print(f"{valor}: {count} empleados ({attrition_percentage[valor]:.2f}%)")

# Gráfico de distribución de Attrition
plt.figure(figsize=(8, 6))

# grafico histograma
attrition_data = pd.DataFrame({
    'Frecuencia': df['Attrition'].value_counts(),
    'Porcentaje': df['Attrition'].value_counts(normalize=True) * 100
})
attrition_data.plot(kind='bar', y='Frecuencia', color=['lightcoral', 'lightgreen'])
plt.title('Distribución de Attrition')
plt.xlabel('Attrition')
plt.ylabel('Frecuencia')

# Añadir etiquetas con valores y porcentajes
for i, v in enumerate(attrition_data['Frecuencia']):
    plt.text(i, v + 20, f'{v}\n({attrition_data["Porcentaje"][i]:.1f}%)', ha='center')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('attrition_distribution.png')

# Analizar variables categóricas destacadas
for col in variables_categoricas:
    if col != 'Attrition':  # Ya analizamos Attrition
        print(f"\nDistribución de {col}:")
        count = df[col].value_counts()
        percent = df[col].value_counts(normalize=True) * 100

        for valor, conteo in count.items():
            print(f"{valor}: {conteo} empleados ({percent[valor]:.2f}%)")

        # Relación con Attrition
        print(f"\nRelación entre {col} y Attrition:")
        tabla = pd.crosstab(df[col], df['Attrition'], normalize='index') * 100
        print(tabla)

        # Histograma para cada categoría
        plt.figure(figsize=(10, 6))

        # Crear un DataFrame para el histograma
        cat_data = pd.DataFrame({
            'Frecuencia': count,
            'Porcentaje': percent
        })
        cat_data.plot(kind='bar', y='Frecuencia', color='skyblue')
        plt.title(f'Distribución de {col}')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')

        # Añadir etiquetas con valores y porcentajes

        for i, v in enumerate(cat_data['Frecuencia']):
            plt.text(i, v + 5, f'{v}\n({cat_data["Porcentaje"][i]:.1f}%)', ha='center')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(f'{col}_distribution.png')

        # Histograma apilado para la relación con Attrition
        plt.figure(figsize=(10, 6))
        cross_tab = pd.crosstab(df[col], df['Attrition'])
        cross_tab.plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral'])
        plt.title(f'Relación entre {col} y Attrition')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')
        plt.xticks(rotation=45)
        plt.legend(title='Attrition')
        plt.tight_layout()
        plt.savefig(f'{col}_vs_attrition.png')
```

```python
# Analisis variables numericas
print("\n--- ANÁLISIS DE VARIABLES NUMÉRICAS ---")

for col in variables_numericas:
    # Estadísticas
    stats = df[col].describe()

    print(f"\nEstadísticas de {col}:")
    print(f"Media: {stats['mean']:.2f}")
    print(f"Mediana: {stats['50%']:.2f}")
    print(f"Desviación estándar: {stats['std']:.2f}")
    print(f"Mínimo: {stats['min']:.2f}")
    print(f"Máximo: {stats['max']:.2f}")

    # Histograma
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    plt.axvline(stats['mean'], color='red', linestyle='--', label=f'Media: {stats["mean"]:.2f}')
    plt.axvline(stats['50%'], color='green', linestyle='--', label=f'Mediana: {stats["50%"]:.2f}')
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'{col}_histograma.png')

    # Histograma separado por Attrition
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=col, hue='Attrition', element='step', common_norm=False,
                 bins=20, palette=['lightgreen', 'lightcoral'])
    plt.title(f'Histograma de {col} por Attrition')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.legend(title='Attrition')
    plt.tight_layout()
    plt.savefig(f'{col}_histograma_attrition.png')

    # Boxplot o grafico de bigotes por Attrition
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Attrition', y=col, data=df, palette=['lightgreen', 'lightcoral'])
    plt.title(f'Diagrama de Caja (Boxplot) de {col} por Attrition')
    plt.xlabel('Attrition')
    plt.ylabel(col)

    # Añadir estadísticas en el gráfico
    for i, attrition in enumerate(['No', 'Yes']):
        subset = df[df['Attrition'] == attrition][col]
        plt.text(i, subset.min() - (subset.max() - subset.min())*0.1,
                 f'Min: {subset.min():.1f}', ha='center')
        plt.text(i, subset.max() + (subset.max() - subset.min())*0.05,
                f'Max: {subset.max():.1f}', ha='center')
        plt.text(i, subset.median(), f'Mediana: {subset.median():.1f}',
                ha='center', va='center', color='white', fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'{col}_boxplot.png')

    # Comparación de medias por Attrition
    mean_yes = df[df['Attrition'] == 'Yes'][col].mean()
    mean_no = df[df['Attrition'] == 'No'][col].mean()
    print(f"Media para Attrition = Yes: {mean_yes:.2f}")
    print(f"Media para Attrition = No: {mean_no:.2f}")
    print(f"Diferencia: {mean_no - mean_yes:.2f}")
```

```python
# Estadisticas descriptivas
print("\n--- ESTADÍSTICAS DESCRIPTIVAS POR GRUPO DE ATTRITION ---")

# Crear tablas descriptivas por grupo
stats_yes = df[df['Attrition'] == 'Yes'][variables_numericas].describe().round(2)
stats_no = df[df['Attrition'] == 'No'][variables_numericas].describe().round(2)

print("\nEstadísticas para empleados que abandonaron (Attrition = Yes):")
print(stats_yes)

print("\nEstadísticas para empleados que permanecieron (Attrition = No):")
print(stats_no)

# Comparación de medianas y medias
print("\nComparación de estadísticas entre grupos:")
for col in variables_numericas:
    media_yes = stats_yes.loc['mean', col]
    media_no = stats_no.loc['mean', col]
    mediana_yes = stats_yes.loc['50%', col]
    mediana_no = stats_no.loc['50%', col]

    print(f"\nVariable: {col}")
    print(f"Media (Yes): {media_yes} | Media (No): {media_no} | Diferencia: {media_no - media_yes:.2f}")
    print(f"Mediana (Yes): {mediana_yes} | Mediana (No): {mediana_no} | Diferencia: {mediana_no - mediana_yes:.2f}")
```

A partir del análisis realizado, se puede observar que aproximadamente un 16% de los empleados deciden dejar la empresa, lo cual representa una cifra significativa.

Durante la exploración de los datos, se identificaron factores clave que influyen en esta decisión. Por ejemplo, trabajar horas extra con frecuencia (OverTime) está altamente relacionado con una mayor tasa de abandono. También se observó que empleados jóvenes, con bajos ingresos mensuales y menor antigüedad en la empresa tienden a renunciar con mayor frecuencia.

La distancia entre el hogar y el lugar de trabajo también parece tener un impacto moderado: mientras más lejos vive el empleado, mayor es la probabilidad de que abandone la organización.

Además, variables como el estado civil y el departamento donde trabaja el empleado mostraron diferencias relevantes, lo que podría estar relacionado con la carga laboral, el ambiente o las oportunidades de desarrollo profesional.

En resumen, los empleados parecen irse por una combinación de factores laborales (como las horas extra y los ingresos) y personales (edad, distancia, estado civil).
